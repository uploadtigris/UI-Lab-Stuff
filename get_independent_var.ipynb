read CSVs

```python
df_newdata = pd.read_csv('/content/annual_aqi_by_county_2023.csv')
df_given = pd.read_csv('/content/Short Paper-Data.csv')
```

```python
pd_data = df_newdata
#Consolidate rows based upon
pd_data['Median AQI'] = pd_data.groupby('State')['Median AQI'].transform('mean')

# Drop duplicates to keep only unique state rows
pd_data = pd_data.drop_duplicates(subset=['State'])
pd_data
```

```python
selected_columns = ['State', 'Median AQI']
pd_data_subset = pd_data[selected_columns].copy()

#Round to three decimal places
pd_data_subset['Median AQI']=pd_data_subset['Median AQI'].round(3)

df_newdata = pd_data_subset

df_newdata
```

Join CSVs

```python
merged_df=pd.merge(df_given, df_newdata, on='State', how='left')

# Drop the "Unnamed: 10" column if it exists
merged_df = merged_df.drop(columns=['Unnamed: 10'], errors='ignore')

#Print which "States" were not in the given dataset
# (Left field) that were in the new dataset
print('These states were unique to the new data and were not included:')
unique_states_right = df_newdata.loc[~df_newdata['State'].isin(df_given['State']), 'State'].unique()
print(unique_states_right)

# Drop the unique rows as they are not in the given dataframe
merged_df = merged_df.head(-3)
```

```python
merged_df
```

```python
# states_to_drop = ['Delaware', 'District of Columbia']
# merged_df = merged_df[~merged_df['State'].isin(states_to_drop)]

states_to_drop = ['Delaware', 'District of Columbia']
merged_df = merged_df[~merged_df['State'].isin(states_to_drop)]
merged_df.reset_index(drop=True, inplace=True)
```
